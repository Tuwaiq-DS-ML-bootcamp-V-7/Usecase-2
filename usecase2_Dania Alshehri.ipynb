## Did any country experience a significant increase or decrease in happiness?

```python
df['Happiness Score % Change'] = df.groupby('Country')['Happiness Score'].pct_change()
sign = df[(np.abs(df['Happiness Score % Change']) >= 0.1)]
print(sign[['Country', 'Happiness Score % Change']])
```

```python
df.sort_values(by=['Country', 'Year'], inplace=True)

# Calculate the change in Happiness Score for each country
df['Happiness Score Change'] = df.groupby('Country')['Happiness Score'].diff()

# Filter the data to find significant changes
# Define a significant change threshold (e.g., 1.0 for substantial increase/decrease)
threshold = 1.0
significant_changes = df[(df['Happiness Score Change'].abs() > threshold)]

# Pivot the data for plotting
pivot_data = significant_changes.pivot(index='Year', columns='Country', values='Happiness Score Change')

# Plot the data
plt.figure(figsize=(12, 6))
pivot_data.plot(kind='bar')
plt.title('Significant Changes in Happiness Scores by Country')
plt.xlabel('Year')
plt.ylabel('Change in Happiness Score')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
# Filter out countries with a happiness score change less than 0.1
filtered_df = df[df['Happiness Score Change'] >= 0.2]

# Visualize the filtered changes using a pandas chart
filtered_df.plot(x='Country', y='Happiness Score Change', kind='barh', figsize=(12, 8), title='Change in Happiness Score by Country')
```

```python
plt.figure(figsize=(12, 8))
for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Happiness Score'], marker='o', label=country)

plt.title('Happiness Score Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
pivot_df = df.pivot(index='Country', columns='Year', values='Happiness Score')

plt.figure(figsize=(20, 8))
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Happiness Score Heatmap by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```
